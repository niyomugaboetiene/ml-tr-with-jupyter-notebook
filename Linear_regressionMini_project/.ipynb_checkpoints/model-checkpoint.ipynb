```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('hours_examscore_data.xlsx')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
# exam score has null values we are going to drop it
df.dropna()
```

```python
df.duplicated().sum() # no duplicate value
```

```python
df.dtypes
```

```python
df['Hours_Studied'] = pd.to_numeric(df['Hours_Studied'], errors='coerce')
```

```python
df['Exam_Score'] = pd.to_numeric(df['Exam_Score'], errors='coerce')
```

```python
df.dtypes
```

```python
df['Hours_Studied'] = df['Hours_Studied'].astype(int)
```

```python
df['Exam_Score'] = df['Exam_Score'].fillna(0).astype(int)
```

```python
df.dtypes
```

```python
# Validate ranges: Hours_Studied ≥ 0 and Exam_Score between 0 and 100.
invalid_hours = df[df.Hours_Studied < 0]
print(invalid_hours)
```

```python
invalid_score = (df[df.Exam_Score < 0]) | (df.Exam_Score > 100)
print(invalid_hours)
```

```python
plt.scatter(df.Hours_Studied, df.Exam_Score, color='green')
plt.xlabel('Hours Studied')
plt.title('Exam Score vs Hours Studied')
plt.ylabel('Exam Score')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['Hours_Studied']], df[['Exam_Score']], test_size=0.2, random_state=42)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_norm = scaler.fit_transform(X_train)
X_test_norm = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()

model.fit(X_train_norm, y_train)
```

```python
model.score(X_test_norm, y_test)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures

model = make_pipeline(PolynomialFeatures(2), LinearRegression())
model.fit(X_train_norm, y_train)
score = model.score(X_test_norm, y_test)
print(score)
```

```python
from sklearn.metrics import mean_absolute_error

y_test_pred = model.predict(X_test_norm)
mean_absolute = mean_absolute_error(y_test, y_test_pred)
print("Mean absolute error", mean_absolute)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared = mean_squared_error(y_test, y_test_pred)
print("Mean squared error", np.sqrt(mean_squared))
```

```python
from sklearn.metrics import r2_score

r2_score_error = r2_score(y_test, y_test_pred)
print("R2", r2_score_error)
```

```python
y_train_pred = model.predict(X_train_norm)

metrics_summary = {
    'Dataset': ['Train', 'Test'],
    'MAE': [
        mean_absolute_error(y_train, y_train_pred),
        mean_absolute_error(y_test, y_test_pred)
    ],
    'RMSE': [
        np.sqrt(mean_absolute_error(y_train, y_train_pred)),
        np.sqrt(mean_absolute_error(y_test, y_test_pred))
    ],
    'R2': [
        r2_score(y_train, y_train_pred),
        r2_score(y_test, y_test_pred)
    ]
}
```

```python
table = pd.DataFrame(metrics_summary)
table.head()
```

```python
plt.scatter(X_test, y_test ,color='red', label='Actual')
plt.plot(X_test, y_test_pred ,color='blue', label='Predicted')
plt.title('Regression line overlaid on the scatter plot')
plt.xlabel('Actual data')
plt.ylabel('Predicted data')
plt.legend()
plt.show()
```

```python
plt.scatter(y_test, y_test_pred, color='green', label='Predicted vs Actual')
plt.title('Predicted vs Actual (Test set)')
plt.xlabel('Actual data')
plt.ylabel('Predicted data')
plt.legend()
plt.show()
```

```python
residuals = y_test - y_test_pred

plt.scatter(y_test_pred, residuals, color='blue', alpha=0.4)
plt.axhline(y=0, color='black', linestyle='--')
plt.title('Residual plo')
plt.xlabel('Predicted Exam Score')
plt.ylabel('Residuals (Actual - Predicted)')
plt.show()
```

```python
import matplotlib.pyplot as plt

residuals = y_test - y_test_pred
plt.hist(residuals, bins=15, color='skyblue', edgecolor='black')
plt.xlabel('Residuals (Actual - Predicted)')
plt.ylabel('Frequency')
plt.title('Residual Distribution (Test set)')
plt.show()
```

```python
K. Short Interpretation Questions (Answer in Markdown)
11. What does the slope (coefficient) mean in the context of study hours and exam scores?
The slope represents the average change in Exam Score for each additional hour studied. 
For example, if the slope is 5, 
it means for every extra hour of study, the exam score increases by about 5 points (on average), 
assuming other factors remain constant.
```

```python
12. Are Train and Test metrics similar? What does that imply?
Train R² = 0.90, Test R² = 0.75
Train errors (MAE, RMSE) are slightly lower than Test errors.
Implication: The model is performing slightly better on training data but is still generalizing reasonably well to unseen data, 
with only mild overfitting.
```

```python
13. Do residuals look randomly scattered around zero? What does that suggest?
Residuals are roughly centered around zero and show no obvious pattern.
Suggestion: This indicates the model is unbiased and captures the main trend. No strong non-linear relationship is left unmodeled 
(though polynomial regression could slightly improve fit).
```
