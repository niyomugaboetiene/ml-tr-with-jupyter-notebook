```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('customer_purchases.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
for col in ['Names', 'Location', 'Blood types', 'Purchased']:
    mode_val = df[col].mode(dropna=True)
    print(mode_val)
    if (len(mode_val)):
         df[col] = df[col].fillna(mode_val.iloc[0])
```

```python
df.isnull().sum()
```

```python
age_mean = df['Age'].mean()
salary_mean = df['Salary'].mean()
print('Age mean', age_mean, 'Salary_mean', salary_mean)
```

```python
df['Age'] = df['Age'].fillna(age_mean)
df['Salary'] = df['Salary'].fillna(salary_mean)
```

```python
df.isnull().sum()
```

```python
df = df.sort_values(['DoB', 'Age'], ascending=[False, False])
```

```python
df['DoB'] = df['DoB'].ffill().bfill()
```

```python
df.isnull().sum()
```

```python
df
```

```python
df['Purchased'] = df['Purchased'].replace({1: 'Yes', 0: 'No'})
```
