```python
import matplotlib.pyplot as plt

weight = [110, 120 , 130, 140, 150]
age = [40, 50, 60, 70, 80]

plt.scatter(weight, age)
plt.xlabel("Weights")
plt.ylabel("age")
plt.show()
```

```python
import matplotlib.pyplot as plt

num1 = [10, 20, 30, 40, 50]
num2 = [1, 2, 3, 4, 5]
plt.plot(num1, num2, color='blue', marker='o', markersize=10, linewidth=3, linestyle='--')
plt.title("Matplotlib visualization")
plt.xlabel("num 1")
plt.ylabel("num 2")
plt.show()
```

```python
import matplotlib.pyplot as plt

num1 = [10, 20, 30, 40, 50]
num2 = [1, 2, 3, 4, 5]
plt.barh(num1, num2, color='blue')
plt.title("Matplotlib visualization")
plt.xlabel("num 1")
plt.ylabel("num 2")
plt.show()
```
