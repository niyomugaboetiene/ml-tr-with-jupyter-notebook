```python
import pandas as pd
import numpy as np

df = pd.read_csv('ds5_customer_survey.csv')
# df.head()
df
```

```python
missing_count = df.isna().sum()
missing_percentage = df.isna().mean()
print("Missing value", missing_count)
print("\nMissing percentage\n", missing_percentage)
```

```python
for col in ["country", "satisfaction", "email"]:
    mode = df[col].mode(dropna=True)
    if len(mode) > 0:
        df[col] = df[col].fillna(mode.iloc[0])
df
```

```python
for col in ["age"]:
    mode = df[col].mode(dropna=True)
    if len(mode) > 0:
        df[col] = df[col].fillna(mode.iloc[0])
df
```

```python
missing_count = df.isna().sum()
missing_percentage = df.isna().mean()
print("Missing value\n", missing_count)
print("\nMissing percentage\n", missing_percentage)
```
