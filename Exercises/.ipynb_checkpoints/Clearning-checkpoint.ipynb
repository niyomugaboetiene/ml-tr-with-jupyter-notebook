```python
import pandas as pd
import numpy as np

df = pd.read_csv('ds1_retail_orders.csv')
# df.head()
df
```

```python
missing_count = df.isna().sum()
missing_percentage = df.isna().mean()
print("Missing value", missing_count)
print("\nMissing percentage\n", missing_percentage)
```

```python
# --- 4.1 Mode imputation (categorical) ---
for c in ["city", "gender"]:
    mode_val = df[c].mode(dropna=True) #ignore missing values
    if len(mode_val) > 0: # fill mode which located on index 0 if there if two or more modes
        df[c] = df[c].fillna(mode_val.iloc[0])
df
```

```python
for col in ["city"]:
    mode = df[col].mode(dropna=True)
    if len(mode) > 0:
        df[col] = df[col].fillna(mode.iloc[0])
df
```

```python
for col in ["unit_price"]:
    mode = df[col].mode(dropna=True)
    if len(mode) > 0:
        df[col] = df[col].fillna(mode.iloc[0])
df
```

```python
missing_count = df.isna().sum()
missing_percentage = df.isna().mean()
print("Missing value\n", missing_count)
print("\nMissing percentage\n", missing_percentage)
```
